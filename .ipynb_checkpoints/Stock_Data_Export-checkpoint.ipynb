```python
#['Symbol', 'Name', 'LastSale', 'MarketCap', 'IPOyear', 'Sector','industry', 'Summary Quote', 'Unnamed: 8']
listings = pd.read_csv('companylist (1).csv')
listings = listings[listings['MarketCap'] != 'n/a']
print(listings.columns)
print(listings.shape)
```

```python
market_cap = listings.MarketCap
market_cap = market_cap[market_cap != 'n/a']
print(market_cap)
```

```python
market_cap = [x.lstrip('$') for x in market_cap]
```

```python
numerical_cap = []
zero_dict = {'M':6,'B':9}

for x in market_cap:
    value = x.lstrip('$').split('.')
    print(value)
    if len(value) > 1:
        letter = value[1][-1]
        value[1] = value[1][:-1]+'0'*(zero_dict[letter]-len(value[1][:-1]))
        value = ''.join(value)
        numerical_cap.append(float(value))
        
    else:
        letter = value[0][-1]
        value = value[0][:-1]+'0'*zero_dict[letter]
        numerical_cap.append(float(value))
```

```python
listings.MarketCap = numerical_cap
```

```python
listings = listings.sort_values('MarketCap',ascending=False)
listings = listings.reset_index(0)
listings = listings[listings.columns[1:]]
```
